```python
import random
import networkx as nx
import matplotlib.pyplot as plt

num_nodes = 15
max_connections = 5
graph = nx.Graph()

for i in range(num_nodes):
    # Düğümlerin mevcut bağlantılarını tutacak bir liste oluşturalım
    connections = random.sample(range(num_nodes), random.randint(1, max_connections))
    for node in connections:
        if i != node:
            weight = random.randint(1, 20)
            graph.add_edge(i, node, weight=weight)

# Grafı çizme
pos = nx.spring_layout(graph)
nx.draw(graph, pos, with_labels=True, node_size=500, node_color='skyblue', font_weight='bold')

# Kenarlara ağırlıkları ekleme
labels = nx.get_edge_attributes(graph, 'weight')
nx.draw_networkx_edge_labels(graph, pos, edge_labels=labels)

plt.title("Her Node'un En Fazla 5 Node İle Bağlantılı Olan Graf")
plt.show()
```

<span style="color:black; font-size:36px">Haversine_distance metodu ile şehirlerin koordinatlarına bakarak aralarındaki mesafeyi hesaplama ve bunu veri setine dönüştürme:</span>

```python
from math import radians, sin, cos, sqrt, atan2

def haversine_distance(lat1, lon1, lat2, lon2):
    R = 6371.0  # Dünya'nın yarıçapı in km

    # Lat ve lon değerlerini radyan cinsine dönüştürme
    lat1 = radians(float(lat1))
    lon1 = radians(float(lon1))
    lat2 = radians(float(lat2))
    lon2 = radians(float(lon2))

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    # Haversine formülü
    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c  # Mesafe hesabı
    return distance
```
